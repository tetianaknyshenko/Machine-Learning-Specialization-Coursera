```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ===== Activation Function Visualization =====

# Define activation functions
def linear(x):
    return x

def logistic(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def swish(x):
    return x / (1 + np.exp(-x))

# Generate input data
x = np.linspace(-10, 10, 500)

# Compute outputs for each activation function
activations = {
    "Linear": linear(x),
    "Logistic (Sigmoid)": logistic(x),
    "ReLU": relu(x),
    "Tanh": tanh(x),
    "Leaky ReLU": leaky_relu(x),
    "Swish": swish(x)
}

# Plot activation functions
plt.figure(figsize=(12, 8))
for i, (name, y) in enumerate(activations.items(), 1):
    plt.subplot(3, 2, i)
    plt.plot(x, y, label=name, color="blue")
    plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
    plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
    plt.title(name)
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid(alpha=0.3)
    plt.legend()

plt.tight_layout()
plt.show()

# Comments:
# 1. Linear: Output grows proportionally to the input; no non-linearity.
# 2. Logistic (Sigmoid): Smoothly maps inputs to (0, 1), useful for probabilities.
# 3. ReLU: Sets negative inputs to 0, widely used in deep learning.
# 4. Tanh: Maps inputs to (-1, 1), useful for centered outputs.
# 5. Leaky ReLU: Similar to ReLU but allows small gradients for negative inputs.
# 6. Swish: Smooth and non-monotonic, often outperforms ReLU in practice.
```
